## Univariate Analysis

```python
## Purpose of univariate analysis to understand the distribution and statistical
## properties of any column/factor
## Mean
## Median
## Mode
## Count
## Range
## Max/Min
## Percentiles
## Outliers
## std/var
```

```python
cont_cols=["temp","atemp","hum","windspeed","casual","registered","cnt"]
cat_cols=["season","yr","mnth","holiday","weekday","workingday","weathersit"]
id_cols=["instant"]
date_cols=["dteday"]
```

## Graphs Used in Univariate Analysis
## Histogram

## Boxplot

# Histogram
#### --> It is used to see bucket-wise/bin-wise distribution of continuous variable

```python
sns.histplot(df["temp"],bins=[0,5,10,15,20,25,30,35,40])
```

# Countplot
#### --> It is used to see the frequency distribution of categorical variable

```python
sns.countplot(df["season"])
```

```python
sns.countplot(df["mnth"])
```

```python
df["season"]=df["season"].map({1:"spring", 2:"summer", 3:"fall", 4:"winter"})
```

```python
for i in cat_cols:
    print("----------------",i,"-------------------")
    sns.countplot(df[i])
    plt.show()
```

```python
## Boxplot
# Quartile distribution of the continuous column
# To see whether the outliers are present in the data
```

```python
sns.boxplot(df["temp"])
```

```python
df["temp"].describe()
```

```python
sns.boxplot(y=df["windspeed"])
```

```python
## Q1=25th percentile (Lower side of the box)
## Q2=50th percentile (Middle of the box) (Median)
## Q3=75th percentile (Upper side of the box)
## IQR=Inter-Quartile Range (Q3-Q1) (75th Percentile - 25th Percentile)
## Lower whisker = Q1-1.5*IQR
## Upper whisker = Q3+1.5*IQR
## Black dots are the outliers
```

```python
arr=np.array([1,2,3,4,5])
```

```python
print("Mean",np.mean(arr))
print("Median",np.median(arr))
```

```python
arr=np.array([1,2,3,4,5,100])
print("Mean",np.mean(arr))
print("Median",np.median(arr))
```
